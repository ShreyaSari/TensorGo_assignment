```python
!pip install pydub SpeechRecognition transformers gtts playsound
!apt-get install ffmpeg

# Step 1: Upload Audio for Voice Capture
from pydub import AudioSegment
from google.colab import files

def upload_voice():
    uploaded = files.upload()  # Upload a .wav file for processing
    for filename in uploaded.keys():
        print(f"Uploaded file: {filename}")
        return filename

audio_filename = upload_voice()

# Step 2: Speech-to-Text
import speech_recognition as sr

def speech_to_text(audio_filename):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio_filename) as source:
        audio = recognizer.record(source)  # Read the audio file
    try:
        text = recognizer.recognize_google(audio)
        print("Recognized Text:", text)
        return text
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand audio")
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")

text_input = speech_to_text(audio_filename)

# Step 3: LLM Integration (GPT-2)
from transformers import pipeline

def get_llm_response(text_input):
    generator = pipeline('text-generation', model='gpt2')
    response = generator(text_input, max_length=100, num_return_sequences=1)
    generated_text = response[0]['generated_text']
    print("LLM Response:", generated_text)
    return generated_text

response_text = get_llm_response(text_input)

# Step 4: Text-to-Speech
from gtts import gTTS

def text_to_speech(text, filename="response.mp3"):
    tts = gTTS(text=text, lang='en')
    tts.save(filename)
    print(f"Response saved as {filename}")

text_to_speech(response_text)

# Step 5: Play Audio
from IPython.display import Audio

def play_audio(filename="response.mp3"):
    return Audio(filename)

play_audio()
```
